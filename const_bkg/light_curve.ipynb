```python
from astropy.time import Time
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt

from pyspi.utils.data_builder.time_series_builder import TimeSeriesBuilderSPI


```

## Plot all dets of a pointing

```python
with fits.open("/home/tguethle/cookbook/SPI_cookbook/examples/automated_Crab/dataset_skymap374-3/spi/evts_det_spec.fits.gz") as hdul:
    hdul.info()
    counts = hdul[1].data["COUNTS"]


counts.shape # (pointings_nr * dets, energy_bins_nr)

with fits.open("/home/tguethle/cookbook/SPI_cookbook/examples/automated_Crab/dataset_skymap374-3/spi/dead_time.fits.gz") as hdul:
    hdul.info()
    time = hdul[1].data['LIVETIME']

time.shape # (pointings_nr * dets, )
```

```python
det = 4
counts_0 = counts[det::19]
total_counts = np.sum(counts_0, axis=1)
time_per_det = time[det::19]
normal_counts = total_counts / time_per_det
plt.plot(normal_counts)
```

### sum over all dets

```python
time[time == 0] = 1
all_E_counts = np.sum(counts, axis=1) / time
print(all_E_counts.shape)
assert all_E_counts.shape[0] % 19 == 0
all_E_counts = all_E_counts.reshape(-1, 19).sum(axis=1)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(all_E_counts)
ax.set_xlabel("Time bzw. Pointing")
ax.set_ylabel("Counts / s (all energies and all detectors)")


ax.grid()
```
